````python
# write a program to iterate name using for loop
name="shivaji"
for char in name:
    print(char)
````

````python
# write a program to find the even numbers in between 100 to 150 using for loop
for n in range(100,151):
    if (n%2==1):
        print(n,end=",")
````

````python
# write a program to read name from list using for loop
names=["vikas","shivaji",'aditya','chandra']
for n in names:
    print("my name is",n)
````
